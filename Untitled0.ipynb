```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([-40,0,40, 80,90, 100, 110, 120, 130, 140, 150, 160])
y = np.array([0,80,160, 170, 180, 190, 200, 210, 220, 230, 240, 250])

plt.figure(figsize=(10,10))

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.plot(x,y)

plt.grid()

plt.show()
```

```python
xhat = np.array([])
yhat = np.array([])

xhat= xhat.astype(int)
yhat= yhat.astype(int)

for i in range(0,x.size):
  xhat= np.append(xhat,(x[i]*1+y[i]*2))
  yhat= np.append(yhat,(y[i]*1))


plt.figure(figsize=(10,10))

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.plot(x,y,'r')
plt.plot(xhat,yhat)
plt.grid()

plt.show()
```

```python
M = np.array([x,y])
T = np.array([[1,2],[0,1]])

C = T.dot(M)

plt.figure(figsize=(10,10))

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.plot(x,y,'r')
plt.plot(C[0],C[1])
plt.grid()

plt.show()
```
